## 0. EDA (Exploratory Data Analysis)

```python
import seaborn as sns

```

## A


#### df.info() kommer att visa information om kolumner, datatyper och antalet icke-saknade värden.

#### df.describe() ger dig en statistisk sammanfattning av numeriska kolumner.
#### df.isnull().sum() visar hur många saknade värden det finns i varje kolumn.
#### df_cleaned är en renad version av datasetet där rader med saknade värden har tagits bort.


```python


# Ladda datasetet
df = sns.load_dataset("mpg")

# Visa de första raderna för att få en översikt
df.head()

# Visa information om datasetet (kolumner, datatyper, saknade värden)
df.info()

# Få en statistisk sammanfattning av datasetet
df.describe()
```

```python
# Kontrollera antalet saknade värden per kolumn
print(df.isnull().sum())

# Ta bort rader med saknade värden (alternativt fyll i med medelvärde/median om det är lämpligt)
df_cleaned = df.dropna()

# Kontrollera att saknade värden har hanterats
print(df_cleaned.isnull().sum())
```

## B

```python
# Välj relevanta numeriska kolumner
relevanta_kolumner = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration']

# Använd describe() på de valda kolumnerna
df_cleaned[relevanta_kolumner].describe()
```

## C

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Skapa en figur med 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Histogram för "mpg"
sns.histplot(df_cleaned['mpg'], kde=True, color='blue', ax=axes[0, 0])
axes[0, 0].set_title('Fördelning av Miles per Gallon (mpg)')
axes[0, 0].set_xlabel('mpg')
axes[0, 0].set_ylabel('Frekvens')

# Histogram för "horsepower"
sns.histplot(df_cleaned['horsepower'], kde=True, color='orange', ax=axes[0, 1])
axes[0, 1].set_title('Fördelning av Hästkrafter (horsepower)')
axes[0, 1].set_xlabel('Hästkrafter')
axes[0, 1].set_ylabel('Frekvens')

# Scatter plot för "weight" vs "mpg" med regressionslinje
sns.regplot(x='weight', y='mpg', data=df_cleaned, color='green', line_kws={'color': 'red'}, ax=axes[1, 0])
axes[1, 0].set_title('Vikt vs Miles per Gallon (mpg) med Regressionslinje')
axes[1, 0].set_xlabel('Vikt')
axes[1, 0].set_ylabel('mpg')

# Scatter plot för "horsepower" vs "mpg" med regressionslinje
sns.regplot(x='horsepower', y='mpg', data=df_cleaned, color='purple', line_kws={'color': 'red'}, ax=axes[1, 1])
axes[1, 1].set_title('Hästkrafter vs Miles per Gallon (mpg) med Regressionslinje')
axes[1, 1].set_xlabel('Hästkrafter')
axes[1, 1].set_ylabel('mpg')

# Justera layouten för att undvika överlapp
plt.tight_layout()
plt.show()
```
